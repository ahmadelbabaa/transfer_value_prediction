## Clustering Players


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

cluster_df = df[['player_name'] + cluster_features].dropna()
X_cluster = StandardScaler().fit_transform(cluster_df[cluster_features])
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_df['cluster'] = kmeans.fit_predict(X_cluster)

pca = PCA(n_components=2)
coords = pca.fit_transform(X_cluster)

plt.scatter(coords[:,0], coords[:,1], c=cluster_df['cluster'])
plt.title('Player Clusters')
plt.show()
```
